```
from sklearn.decomposition import PCA
```

```
csv_url = 'https://raw.githubusercontent.com/riddlemeS4m/machine-learning-scientist-datacamp/dev-google/module-3-unsupervised-learning/wine.csv'
df = pd.read_csv(csv_url)
```

```
df.keys()
```

```
len(df)
```

```
samples = df[['total_phenols','od280']].values
samples[:5]
```

```
# samples is an array of two features

model = PCA()
model.fit(samples)
transformed = model.transform(samples)

# new array has same number of rows and features as the sample
```

```
print(transformed[:2],transformed[177])
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(samples[:,0],samples[:,1],c=df['class_label'])
# plt.axes('equal')
plt.xlabel('total_phenols')
plt.ylabel('od280')
plt.axvline(x=0,color='black')
plt.axhline(y=0,color='black')
plt.xlim(-4,6)
plt.ylim(-4,6)
plt.show()
```

```
plt.scatter(transformed[:,0],transformed[:,1],c=df['class_label'])
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```

```
# PCA features are not linearly correlated
# for example, even though BMI and weight are directly correlated, reducing these features with PCA would take out the correlation
```

```
model.components_
# PCA learns the principal components of features (strongest direction of variance) and uses that to transform the data
```

```
from scipy.stats import pearsonr
```

```
correlation, pvalue = pearsonr(samples[:,0],samples[:,1])
print(correlation,pvalue)
```
